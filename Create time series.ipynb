```python
import numpy as np
import pandas as pd
```

```python
cal = pd.read_csv('calendar.csv')
```

```python
cal.head()
```

```python
price = pd.read_csv('sell_prices.csv')
```

```python
price.head()
```

```python
sales_train_validation = pd.read_csv('sales_train_validation.csv')
```

```python
sales_train_validation.head()
```

```python
for i in range(2):#(sales_train_validation.shape[0]): #uncomment here to create all datafiles
    name = sales_train_validation.loc[i,'id']
    
    store = sales_train_validation.loc[i,'store_id']
    item = sales_train_validation.loc[i,'item_id']
    item_price = price[(price.store_id == store) & (price.item_id == item)][['wm_yr_wk','sell_price']].copy()
    
    item_sales = sales_train_validation.iloc[i,6:].reset_index().copy()
    item_sales = pd.DataFrame(item_sales.values, columns= ['d','units_sold'])
    
    add_cal = pd.merge(cal, item_sales, how='right', on='d')
    add_prices = pd.merge(add_cal, item_price, how='left', on='wm_yr_wk')
    add_prices.reset_index(drop=True,inplace=True)
    add_prices = add_prices.set_index('date')
    add_prices.to_csv('./data/{}.csv'.format(name))
```
